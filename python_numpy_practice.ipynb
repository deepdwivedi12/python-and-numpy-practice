```python
import pandas as pd

# Creating DataFrame
data = {
    "Name": ["Deep", "Rahul", "Aman", "Riya", "Priya"],
    "Department": ["IT", "IT", "HR", "HR", "IT"],
    "Marks": [85, 90, 78, 88, 95]
}

df = pd.DataFrame(data)

print("Original DataFrame")
print(df)

# Filtering
print("\nStudents having Marks > 85")
filtered_df = df[df["Marks"] > 85]
print(filtered_df)

# Sorting
print("\nSorting by Marks")
sorted_df = df.sort_values(by="Marks", ascending=False)
print(sorted_df)

# Grouping
print("\nGroup By Department")
grouped = df.groupby("Department")
print(grouped["Marks"].sum())

# Aggregation
print("\nAggregation Functions")
result = df.groupby("Department").agg({
    "Marks": ["sum", "mean", "max", "min", "count"]
})

print(result)
```
